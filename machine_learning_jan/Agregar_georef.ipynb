```python
import pandas as pd
```

```python
df1 = pd.read_csv("CONSOLIDADO.csv", index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv("TiempoDistancia.csv")
```

```python
df2.head()
```

```python
df1["tiempov"] = df1["codigo_x"].map(df2.set_index("id_trabajador")["TiempoViaje"].to_dict())
df1["dist"] = df1["codigo_x"].map(df2.set_index("id_trabajador")["Distancia"].to_dict())
df1["GMU"] = df1["codigo_x"].map(df2.set_index("id_trabajador")["GMU2010"].to_dict())
df1["IMU"] = df1["codigo_x"].map(df2.set_index("id_trabajador")["IMU2010"].to_dict())
```

```python
df1.head()
```

```python
df1.isna().values.any()
```

```python
missing_values = df1[df1.isna().any(axis=1)]
```

```python
df1 = df1.dropna()
```
