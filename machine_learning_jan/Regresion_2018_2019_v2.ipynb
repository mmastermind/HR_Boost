```python
import os
import pandas as pd
```

```python
df5 = pd.read_csv("CONSOLIDADO_DESDE_2018_v2.csv", index_col=0)
```

```python
df5.head()
```

```python
df5 = df5[["dias_pag","dias_lab","sueldo","comp","asist","transp","vales_x","te_dob",
          "te_trip","desc_Lab","fest_lab","dominic","perc_grav","obj","t_ded","label"]]
```

```python
df5.head()
```

```python
X = df5.drop("label", axis=1)
y = df5["label"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
model1 = pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```
