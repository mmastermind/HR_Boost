```python
import os
import re
import pandas as pd
```

```python
data = []

for file in os.listdir("PERCEPCIONES_2019"):
    if file.endswith(".txt"): 
        with open(os.path.join("PERCEPCIONES_2019", file)) as f:
            week = file[17:19]
            for line in f.readlines():
                if len(line) == 253:
                    line = re.sub("\s+", "|", line.strip())
                    line = re.sub(",", "", line.strip()) #esto quita las comas de las cifras
                    line = f"{week}|{line}"
                    if not line:
                        continue
                    line_list = [s for s in line.split(';') if s]
                    if not line_list:
                        continue
                    data.append(line_list)
    else:
        continue
```
